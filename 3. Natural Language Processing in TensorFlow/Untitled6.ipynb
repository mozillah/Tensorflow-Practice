```python
import tensorflow_datasets as tfds
```

```python
data,info=tfds.load("imdb_reviews",with_info=True,as_supervised=True)
```

```python
train_data,test_data=data["train"],data["test"]
```

```python
training_sentences = []
training_labels = []
```

```python

testing_sentences = []
testing_labels = []
```

```python
training_labels = []
training_sentences=[]
for sentence,label in train_data:
    training_sentences.append(sentence.numpy().decode("utf8"))
    training_labels.append(label.numpy())
```

```python
testing_sentences = []
testing_labels=[]
for sentence,label in test_data:
    testing_sentences.append(sentence.numpy().decode("utf8"))
    testing_labels.append(label.numpy())
```

```python
vocab_size = 10000
embedding_dim = 16
max_length = 120
trunc_type='post'
oov_tok = "<OOV>"
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)
```

```python
# training_sentences
```

```python
word_index = tokenizer.word_index
sequences =tokenizer.texts_to_sequences(word_index)
padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type)
```

```python

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences,maxlen=max_length)
```

```python
lists=[]
for key,value in word_index.items():
    print(key,value)
    lists.append((value,key))
```

```python
reverse_word_index=dict(lists)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Embedding(vocab_size,embedding_dim,input_length=max_length)
                             ,tf.keras.layers.Flatten(),
                             tf.keras.layers.Dense(128,activation="relu")
                             ,tf.keras.layers.Dense(1,activation="sigmoid")])
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=["accuracy"])
```

```python
num_epochs = 10
model.fit(padded, training_labels, epochs=num_epochs)
```
