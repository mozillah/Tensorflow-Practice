```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Simple Neural Network with one neuron as input

`Dense` is used to define a layer of neuron. One Dense means one layer in it 

 and `unit`=1 means one neuron in it. 
 
 `Successive layers` defined in Sequence so we use Sequential
 
 `input_shape` means shape of input to our network

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Neural network => No idea of relationship b/w **X** and **Y** so makes guess
after guess it checks with actual data whether guess is good or not
`Loss function` measures how good our guess is.
After guess `loss function` gives data to optimizer and optimizer check how badly the guesss was done.
Each guess should be better than before guess.
Accuracy => Improves and reaches to 100 means convergence.

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```
