```python
nrows,ncols=10,10
pic_ind = 0
fig=plt.gcf()
fig.set_size_inches(4*nrows,ncols*4)
horse_pix_8 = [os.path.join(train_horse_dir,name) for name in train_horse_names[pic_ind:+2*nrows]]
human_pix_8 = [os.path.join(train_human_dir,name) for name in train_human_names[pic_ind:2*ncols]]

for ind,name in enumerate(horse_pix_8+human_pix_8):
    sp= plt.subplot(nrows,ncols,ind+1)
    img = cv2.imread(name)
    plt.imshow(img[:,:,::-1])
    
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16,(3,3),activation='relu',input_shape=(300,300,3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(32,(3,3),activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Flatten(),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    # Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('horses') and 1 for the other ('humans')
    tf.keras.layers.Dense(1, activation='sigmoid')
    
])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop,Adagrad
```

```python
model.compile(optimizer=RMSprop(lr=0.001),loss="binary_crossentropy",metrics=["accuracy"])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
trainGen = ImageDataGenerator(rescale=1/255)
```

```python
train_generator =trainGen.flow_from_directory("/home/mazz/Desktop/github/tmp/horse-or-human/",target_size=(300,300),batch_size=128,class_mode='binary')
```

```python
history = model.fit(
      train_generator,
      steps_per_epoch=8,  
      epochs=15,
      verbose=1)
```
