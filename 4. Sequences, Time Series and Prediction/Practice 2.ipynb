```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
def trend(time, slope=0):
    return slope * time

def plot_series(time, series, format="-", start=0, end=None, label=None):
    plt.plot(time[start:end], series[start:end], format, label=label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
```

```python
time = np.arange(4 * 365 + 1)
baseline = 10
series = trend(time, -0.3)
plot_series(time,series)
series = trend(time, 0.3)
plot_series(time,series)
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(series)
```

```python
def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
baseline = 10
amplitude = 40
series = seasonality(time, period=365, amplitude=amplitude)
plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
slope = -0.05
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)
plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
def white_noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
noise_level = 5
noise = white_noise(time, noise_level, seed=42)

plt.figure(figsize=(10, 6))
plot_series(time, noise)
plt.show()
```

```python
series += noise

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
def impulses(time,impulses,amplitude=10,seed=None):
    rnd =np.random.RandomState(seed)
#     print(rnd)
    impulse_indices = rnd.randint(len(time),size=10)
    ar = np.zeros(len(time))
#     print(ar)
    for ind in impulse_indices:
#         ar[ind] = rnd.randn()
        ar[ind] += rnd.rand() * amplitude
#     print(ar)
    return ar
```

```python
def autocorrelation(source,ds):
    ar = source.copy()
    max_lag = len(ds)
    for step,value in enumerate(source):
        for lag,d in ds.items():
            if step-lag >0:
                ar[step] += d* ar[step-lag]
    return ar
```

```python
signal = impulses(time,30)
series = autocorrelation(signal,{30:0.9})
plot_series(time,series)
plt.plot(time, signal, "k-")
```
