```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_series(time,series):
    plt.figure(figsize=(10,8))
    plt.plot(time,series)
    plt.xlabel("time")
    plt.ylabel("series")
    plt.grid(True)
    plt.show()
```

```python
time = np.arange(3*365+2)
time
```

```python
def trend(time,slope=0.1):
    series = time * slope
    return series
```

```python
series = trend(time,slope=0.1)
```

```python
series
```

```python
plot_series(time,series)
```

```python
def seasonality_pattern(season_time):
    return np.where(season_time<0.4,
                   np.cos(season_time),
                    np.sin(season_time)+np.tan(season_time)
                   
                   )

def seasonality(time,period,amplitude,phase=0):
    seasons=((time + phase)%period)/period
    return (amplitude * seasonality_pattern(seasons))
```

```python
def seasonality_pattern(season_time):
    return np.where(season_time<0.4,
                   2*np.cos(season_time*np.pi),
                    2*(np.sin(season_time)+np.tan(season_time))
                   )

def seasonality(time,period,amplitude,phase=0):
    seasons=((time + phase)%period)/period
    return (amplitude * seasonality_pattern(seasons))
```

```python
series = seasonality(time,365,10)
plot_series(time,series)
```

```python
baseline = 5
amplitude = 10

slope = 0.05
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)
plot_series(time, series)
```
